# Google Search Analysis

Google Trends can be used for the task of Google search analysis. Google Trends provides an API that can be used to analyze the daily searches on Google. This API is known as pytrends, you can easily install it in your systems by using the pip command; pip install pytrends.

```python
pip install pytrends
```

Now let’s get started with the task of Google search analysis by importing the necessary Python libraries:

```python
#install required libraries
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends= TrendReq()
```

Here we will be analyzing the Google search trends on the queries based on “Data Analytics”, so let’s create a DataFrame of the top 10 countries which search for “Data Analytics” on Google

```python
#words most searched by 
trends.build_payload(kw_list=["Data Analytics"])
data=trends.interest_by_region()
data=data.sort_values(by="Data Analytics", ascending=False)
data=data.head(10)
print(data)
```

So, according to the above results, the search queries based on “Data Analytics” are mostly done in Singapore. We can also visualize this data using a bar chart

```python
#data visualization
data.reset_index().plot(x='geoName', y="Data Analytics", 
                         figsize=(20,20), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

So as we all know that Data Analytics has been the focus of so many companies and students for the last 3-4 years, so let’s have a look at the trend of searches to see how the total search queries based on “Data Analytics” increased or decreased on Google

```python
data=TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=["Data Analytics"])
data=data.interest_over_time()
fig, ax=plt.subplots(figsize=(20,15))
data['Data Analytics'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Data Analytics', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```

# Conclusion
So we can see that searches based on “Data Analytics” on Google started to increase in 2017 and the highest searches were done in 2020 till today. This is how we can analyze Google searches based on any keyword. A business can perform Google search analysis to understand what people are looking for on Google at any given time.
